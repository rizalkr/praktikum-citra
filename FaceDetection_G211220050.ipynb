```python
import glob
import zipfile
import os
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# ==========================================
# 1. KONFIGURASI NAMA FILE
# ==========================================
zip_database = "db-face.zip"      # File ZIP database wajah
zip_test     = "folder_test.zip"  # File ZIP foto tes

# Nama folder tempat kita akan menaruh hasil ekstrak (Bebas, saya namai agar rapi)
folder_db_extracted   = "db_face_hasil_unzip"
folder_test_extracted = "folder_test_hasil_unzip"

# ==========================================
# 2. FUNGSI UNTUK EKSTRAK ZIP
# ==========================================
def ekstrak_zip(nama_zip, folder_tujuan):
    print(f"Mengekstrak '{nama_zip}' ke '{folder_tujuan}'...")
    if not os.path.exists(folder_tujuan):
        os.makedirs(folder_tujuan)

    try:
        with zipfile.ZipFile(nama_zip, 'r') as zip_ref:
            zip_ref.extractall(folder_tujuan)
        print(f"Berhasil diekstrak!")
    except FileNotFoundError:
        print(f"ERROR: File '{nama_zip}' tidak ditemukan! Pastikan nama file benar.")
        return False
    return True

# Ekstrak kedua file
if not ekstrak_zip(zip_database, folder_db_extracted):
    raise SystemExit("Stop: Database zip tidak ditemukan.")
if not ekstrak_zip(zip_test, folder_test_extracted):
    raise SystemExit("Stop: Folder test zip tidak ditemukan.")

# ==========================================
# 3. CARI SEMUA FILE FOTO (RECURSIVE)
# ==========================================
# Kita cari sampai ke folder terdalam (recursive=True) jaga-jaga ada folder dalam folder
print("\nMencari file gambar...")
daftar_file = glob.glob(f"{folder_test_extracted}/**/*.jpg", recursive=True)
daftar_file.extend(glob.glob(f"{folder_test_extracted}/**/*.png", recursive=True))
daftar_file.extend(glob.glob(f"{folder_test_extracted}/**/*.jpeg", recursive=True))

print(f"Ditemukan {len(daftar_file)} file foto untuk diproses.")
```

```python
# === LANGKAH 2: Loop melalui setiap file ===
for file_foto_uji in daftar_file:

    # Cetak file yang sedang diproses
    print(f"\n======================================")
    print(f"MEMPROSES: {file_foto_uji}")
    print(f"======================================")

    # Reset variabel untuk setiap foto baru
    nama_terdeteksi = "Tidak Dikenal"
    umur_terdeteksi = "Tidak Diketahui"
    analisis = [] # kosongkan hasil analisis sebelumnya

    # -----------------------------------------------------------
    # BAGIAN 1: IDENTIFIKASI NAMA
    # -----------------------------------------------------------
    try:
        df_list = DeepFace.find(
            img_path=file_foto_uji,
            db_path=folder_database,
            model_name="VGG-Face",
            enforce_detection=False
        )

        if len(df_list) > 0 and not df_list[0].empty:
            hasil_teratas = df_list[0].iloc[0]
            nama_folder = hasil_teratas['identity'].split('/')[-2]
            nama_terdeteksi = nama_folder
            print(f"--- Hasil Identifikasi (NAMA): {nama_terdeteksi}")
        else:
            print("--- Hasil Identifikasi (NAMA): Wajah tidak ditemukan di database.")

    except Exception as e:
        print(f"Error saat identifikasi nama: {e}")

    # -----------------------------------------------------------
    # BAGIAN 2: PERKIRAAN UMUR
    # -----------------------------------------------------------
    try:
        analisis = DeepFace.analyze(
            img_path=file_foto_uji, # Menggunakan file dari loop
            actions=['age'],
            enforce_detection=True
        )

        hasil_wajah_pertama = analisis[0]
        umur = hasil_wajah_pertama['age']
        umur_terdeteksi = f"{umur} tahun"

        print(f"--- Hasil Analisis (UMUR): Perkiraan {umur} tahun")

    except Exception as e:
        print(f"Error saat analisis umur: {e}")


    # -----------------------------------------------------------
    # OPSIONAL: Menampilkan hasil gabungan
    # -----------------------------------------------------------
    # (Kode ini akan menampilkan gambar satu per satu di output Colab)
    if 'nama_terdeteksi' in locals() and 'umur_terdeteksi' in locals():
        try:
            img = cv2.imread(file_foto_uji)

            if len(analisis) > 0:
                region = analisis[0]['region']
                x, y, w, h = region['x'], region['y'], region['w'], region['h']

                # --- Pengaturan Padding (dari diskusi kita sebelumnya) ---
                padding = 20
                x_pad = max(0, x - padding)
                y_pad = max(0, y - padding)
                w_pad = w + (padding * 2)
                h_pad = h + (padding * 2)
                # --- Selesai Padding ---

                cv2.rectangle(img, (x_pad, y_pad), (x_pad + w_pad, y_pad + h_pad), (0, 255, 0), 2)

                label = f"{nama_terdeteksi} ({umur_terdeteksi})"
                text_x = x_pad
                text_y = y_pad - 10

                if text_y < 25:
                    text_y = y_pad + h_pad + 25

                cv2.putText(img, label, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            plt.figure(figsize=(10, 8))
            plt.imshow(img_rgb)
            # Judul plot diubah agar dinamis
            plt.title(f"Hasil Deteksi untuk: {file_foto_uji}", fontsize=14)
            plt.axis('off')
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"\nTidak bisa menggambar kotak atau menampilkan output: {e}")
    else:
        print("\nTidak dapat menampilkan gambar karena identifikasi/analisis gagal.")

print("\n--- PROSES BATCH SELESAI ---")
```
